```python
import tensorflow as tf
```

```python
tensor_one_d = tf.constant([1,2,3,4,5])
print(tensor_one_d)
```

```python
tensor_two_d = tf.constant([[1,2,3],[4,5,6]])
print(tensor_two_d)
```

```python
tensor_three_d = tf.constant([
    [[1,2,3],
     [4,5,6]],
    [[7,8,9],
     [3,4,5]]
])
print(tensor_three_d)
```

Converting numpy array to tensor

```python
import numpy as np
```

```python
np_array = np.array([1,2,3,4,5])
tensor_array = tf.convert_to_tensor(np_array)
print(tensor_array)
```

Identity matrix

```python
i_matrix = tf.eye(
    num_rows = 3,
    num_columns = 3,
    batch_shape = None,
    dtype = tf.dtypes.float32,
    name = None
)
print(i_matrix)
```

```python
i_matrix = tf.eye(
    num_rows = 3,
    num_columns = 3,
    batch_shape = [2,],
    dtype = tf.dtypes.float32,
    name = None
)
print(i_matrix)
```
